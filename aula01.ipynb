Do resultados anterior, podemos concluir que alguns dos imóveis têm preços com os rótulos "Mês", "Ano" e "Dia" associado, o que indica que talvez tenham sido precificados por seus valores de aluguéis por dia, mês e ano, ao contrário dos demais, rotulados como "None".

Agora, vamos incluir essas informações no nosso DataFrame original, para que possamos ter uma visão melhor.

### Incluir as colunas do DataFrame auxiliar, no DataFrame original

```python
dados[['Moeda', 'Valor_anuncio', 'Tipo_anuncio']] = dados['Valor'].str.split(expand=True)
dados.head()
```

```python
# verificar imóveis em que 'Tipo_anuncio' é 'None', ou seja, os imóveis que foram precificado sem base em dia/mês/ano.
dados_filtrados = dados[dados["Tipo_anuncio"].isnull()]
dados_filtrados
```

### Transformar 'Valor_anuncio' do tipo 'object' para 'float'

O objetivo da manipulação é possibilitar a manipulação dos dados que estam em formato incorreto. 

Originalmente, os dados de valor eram do tipo 'object', quando na verdade representavam valores numérico, ou seja, deveriam ser 'float'. Para isso, seguimos os seguintes passos:

- Dividimos os elementos de 'Valor' em uma lista, com o método split()
- Cada elemento da lista virou uma coluna de um DataFrame, nomeados como [Moeda, Valor_anuncio, Tipo_anuncio]
- Inserimos o novo DataFrame no DataFrame original
- Filtramos os imóveis em que o 'Tipo_anuncio' é 'None', ou seja, o valor do imóvel não é associado ao rótulo de dia/mês/ano
- Desses imóveis filtrados, vamos transformar seu valor no tipo float (passos a seguir)


```python
# antes
dados_filtrados.info()
```

```python
# antes
dados_filtrados.head()
```

Converter a string de 'Valor_anuncio': retirando o '.' e transformando em float. 

É necessário retirar o '.', uma vez que foi utilizado para demarcar os milhares e não aos decimais, o que causa erro


```python
dados_filtrados['Valor_anuncio'].str.replace('.','').astype(float)
```

```python
# substituir os dados convertidos no data frame original
dados_filtrados['Valor_anuncio'] = dados_filtrados['Valor_anuncio'].str.replace('.','').astype(float)
```

```python
# depois
dados_filtrados
```

```python
#depois
dados_filtrados.info()
```

Nota-se que o Dtype da coluna 'Valor_anuncio' passou de 'object' para 'float64'

## Calcular o preço do m² do imóvel com base no valor do anúncio e da metragem

Aqui, calculamos o preço do m² de cada imóvel, independente de sua localização. O que pode acontecer é que imóveis discrepantes (outliers) podem nos das conclusões diferentes do padrão esperado, já que o preço depende apenas das informações de metragem e valor associadas ao imóvel.

```python
dados_filtrados['Valor_m2'] = dados_filtrados['Valor_anuncio']/dados_filtrados['Metragem']
dados_filtrados.head()
```

## Calcular o preço do m² dos imóveis segundo o bairro, baseado na quantidade de imóveis

Nesse caso, o preço do m² do bairro é calculado considerando uma média de médias. Isso acontece porque inicialmente calculamos o valor do m² para cada imóvel, dividindo o seu valor pela sua metragem ('Valor_m2'). Em seguida, quando ocorre um agrupamento por bairros (groupby('Bairro')), **esses valores de m² são somados e divididos pela quantidade de imóveis**, ou seja, não indica necessariamente o valor do m² segundo sua metragem, mas sim uma média baseado no número de imóveis.


```python
dados_filtrados.head()
```

```python
dados_filtrados.groupby('Bairro').mean()
```

## Calcular o preço do m² dos imóveis segundo o bairro em que está localizado, baseado na metragem total.

Para isso, os bairros são agrupados, calculando o somatório do 'Valor_m2' dos imóveis que pertencem a cada bairro, para posterior cálculo de média. 

Dessa forma, podemos calcular uma média baseado no **somatório geral de metragem e valor dos imóveis**. Assim, podemos ter uma visão mais ampla do valor do m², do que aquela que depende somente da média baseado no preço de um imóvel isolado.

```python
print('Agrupar os imóveis dos bairros em uma nova variável, calculando a soma das informações disponíveis, para posteriormente calcular a média\n')
dados_bairro = dados_filtrados.groupby('Bairro').sum()
dados_bairro
```

```python
print('Cálculo do valor do m² dos bairros (somatório dos valores dos imóveis daquele bairro dividido pelo somatório das metragem dos imóveis naquele bairro) \n')
dados_bairro['Valor_m2_bairro'] = dados_bairro['Valor_anuncio']/dados_bairro['Metragem']
dados_bairro
```

## Vamos comparar o preço do m² obtidos nas duas abordagens

Aqui, vamos considerar o exemplo do bairro da Vila Mariana. Será que a forma com que foram calculados esses valores muda o resultado obtido?

```python
m2_n_imoveis = dados_filtrados.groupby('Bairro').mean().loc['Vila Mariana']['Valor_m2']
print(f'Valor do m² no bairro Vila Mariana, baseado na quantidade de imóveis: {m2_n_imoveis}')
```

```python
m2_metragem_total = dados_bairro.loc["Vila Mariana"]['Valor_m2_bairro']
print(f'Valor do m² no bairro Vila Mariana, baseado na metragem total: {m2_metragem_total}')
```

```python
print(f'A diferença no valor do m² no bairro Vila Mariana, segundo as duas abordagens, foi de R${m2_n_imoveis-m2_metragem_total:.2f}')
```
